Loading Titanic dataset from seaborn


```python
df = sns.load_dataset('titanic')
df
```

```python
df.info()
```

```python
df.isnull().sum()  # This will shows the null values of dataset
```

```python
avg_age = df['age'].mean()
print(f'Average age: {avg_age}')
```

```python
avg_age = round(df['age'].mean()) # This will make avg as round off value
print(f'Average age: {avg_age}')
```

```python
df.fillna
```

```python
df['age'].fillna(avg_age, inplace = True)
df.head(20)
```

```python
# drop null values in 'embarked' column # it will delete the mssing values of the columns
df.dropna(subset = ['embarked'], inplace= True)
```
