```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic') 
df
```

```python
#histogram of age distribution # for single data or column

fig = sns.histplot(df, x='age', kde = True, bins = 30, color= 'blue', edgecolor = 'white') #kde- kernal density estimate
fig.set_title("Age Distribution of Titanic Passengers")
fig.set_xlabel("Age")
fig.set_ylabel("Frequency")
```

```python
# countplot : survival based on gender

fig = sns.countplot(df, x= "survived", palette = "RdBu", hue = "sex") # hue : defines category, palette: theme colors 
fig.set_title("Survival count  of Titanic Passengers")
fig.set_xlabel("Survival")
fig.set_ylabel("Count")
```

```python
# we use barplot when we have more than one coloumn of datset

sns.barplot(df, x='sex', y= 'fare', hue ='class') 
fig.set_title("Average Fare by Passengers Class and Gender")
fig.set_xlabel("sex")
fig.set_ylabel("Average Fare")
```

```python
fig = sns.countplot(df, x= 'class', palette = 'viridis')
fig.set_title("Count of Passenger by Class")
fig.set_xlabel(" Passenger class")
fig.set_ylabel("Count")
```

```python
# we use pair plot function to show mutiple graphs and their relation with each other

sns.pairplot(df[['age', 'fare', 'survived', 'pclass']], hue = 'survived',  palette = 'terrain')
```
