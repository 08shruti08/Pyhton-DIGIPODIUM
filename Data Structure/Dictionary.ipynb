```python
# to give info into the dictionary by "dict()"
info = dict(name= "Shruti", age = 23, city = "New York", gender = "Female")
print(info)
```

```python
student = {
    "name" : "John",
    "age" : 20,
    "class" : 12,
    "gender" : input("Enter your gender: ") # user input in dictionary
}
print(student)
```

```python
# 1st method to acccess the value of a key
print(info['name'])
```

```python
# 2nd method to acccess the value of a key
print(info.get('age'))
```

```python
#to add new pair to the dictionary
info['country'] = 'USA'
print (info)
```

```python
info["class"] = 9
print(info)
```

```python
#to check attributes of the dictionary

print(info.keys())
print(info.values())
print(info.items()) #for both keys:values
```

```python
for k,v in info.items():
    print(f"{k} : {v}")
```

```python
#Nested Dictionary

company = {
    'emp001' : {
        'name' : 'Shruti',
        'age' : 23,
        'department' : 'IT'
    },
    'emp002' : {
        'name' : 'alice',
        'age' : 26,
        'department' : 'HR'
    },
     'emp003' : {
        'name' : 'Bob',
        'age' : 29,
        'department' : 'Finance'
    },
}
print(company)
```

```python
# to print dictionary horizontally coz default in dict(keys : values) print vertically

from pprint import pprint
pprint(company)
```

```python
# to access values in nested dictionary

company['emp001']['age']
```

```python
company['emp004']={
     'name' : 'varun',
        'age' : 22,
        'department' : 'Electrinocs'
}
print(company)
```
